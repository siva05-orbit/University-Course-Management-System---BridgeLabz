```python
def grading_engine(grades_raw):
    splitted= grades_raw.split()
    my_dict={}
    for item in splitted:
        name=item.split(":")[0]
        code=item.split(":")[1]
        mark=item.split(":")[2]

        if name not in my_dict:
            my_dict[name]={}
            my_dict[name][code]=mark
        else:
            my_dict[name][code]=mark
    return my_dict


grading_engine("alice:CS101:88 alice:CS201:76 alice:EN101:92 bob:CS101:65 charlie:CS201:55 charlie:MA101:48 charlie:EN101:71")


```
